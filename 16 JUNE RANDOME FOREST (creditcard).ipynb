```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score,confusion_matrix,accuracy_score

from pylab import rcParams
rcParams['figure.figsize']=8,8 #used for having same graphs size for all graphs in this
```

```python
data=pd.read_csv("creditcard.csv")
```

```python
x=data.drop(columns="Target")
y=data["Target"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0,stratify=y)
```

```python
random=RandomForestClassifier(criterion='entropy',max_depth=4,n_estimators=80)
random.fit(x_train,y_train)
```

```python
y_pred=random.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
accuracy_score(y_pred,y_test)
```

```python
roc_auc_score(y_pred,y_test)
```

```python
y_pred_train=random.predict(x_train) #checking for overfitting
```

```python
roc_auc_score(y_pred_train,y_train)
```
